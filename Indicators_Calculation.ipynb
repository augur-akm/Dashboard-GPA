```python
import pandas as pd
```

# M5

```python
data = pd.read_excel("df_M5.xlsx", dtype={"tender.procuringEntity.identifier.id":str}, parse_dates=["tender.date", "tender.tenderPeriod.endDate"])

print("The shape of the DF: ", data.shape)
data.head(3)
```

## M51

```python
df = pd.pivot_table(data=data, values = "tender.value.amount_SDR", 
                    index=data["tender.num_lots"]!=1, columns=data["tender.date"].dt.year, 
                    aggfunc="count")

pd.concat([df, df/df.sum()], axis=1)
```

```python
df = pd.pivot_table(data=data, values = "tender.value.amount_SDR", 
                    index=data["tender.num_lots"]!=1, columns=data["tender.date"].dt.year, 
                    aggfunc="sum")

pd.concat([df, df/df.sum()], axis=1)
```

```python
pd.concat([df, df/df.sum()], axis=1).sum()
```

## M57

```python
data_short = data[data["tender.procurementMethod"]=="open"]
data_short["tender.tenderPeriod.endDate"] = pd.to_datetime(data_short["tender.tenderPeriod.endDate"])
data_short["duration"] = (data_short["tender.tenderPeriod.endDate"] - data_short["tender.date"]).dt.days

data_short = data_short[data_short["duration"]>=0]#!!!!!!!!!!!!!!!!!!!!!!!!!!!1

print("The shape of the DF: ", data_short.shape)
data_short.head(3)
```

```python
data_short["duration"].groupby(data_short["tender.date"].dt.year).agg(["median", "mean"]).T
```

```python
df = pd.pivot_table(data=data_short, values = "tender.value.amount_SDR", 
                    index=data_short["duration"]<40, columns=data_short["tender.date"].dt.year, 
                    aggfunc="count")

pd.concat([df, df/df.sum()], axis=1)
```

```python
df = pd.pivot_table(data=data_short, values = "tender.value.amount_SDR", 
                    index=data_short["duration"]<40, columns=data_short["tender.date"].dt.year, 
                    aggfunc="sum")

pd.concat([df, df/df.sum()], axis=1)
```
